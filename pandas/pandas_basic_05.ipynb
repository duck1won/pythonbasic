```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
def change_boolean(value):
    if "yes" in value:
        return True
    else:
        return False

titanic["alive"] = titanic["alive"].apply(change_boolean)
titanic
```

```python
titanic["age"].isnull()
titanic["age"].isnull().sum()
titanic["age"] = titanic["age"].dropna()
titanic["adult_female"] = titanic.apply(lambda r : True if r.age>=20 and r.sex == "female" else False,axis=1)
titanic
```
