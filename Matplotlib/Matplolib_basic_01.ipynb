```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi * x1) * np.exp(-x1)
y2 = np.cos(2*np.pi * x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,"yo-")
plt.title("A tale of 2 subplots")
plt.ylabel("Damped oscillation")

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,"r.-")
plt.xlabel("time (s)")
plt.ylabel("Undamped")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
x = titanic['alive'].value_counts()
people=['사망자','생존자']
colors=["red","green"]
plt.title("타이타닉호 생존 여부")
plt.barh(people, x,color=colors)
plt.xlabel("인원 수")
plt.show()
```

```python
labels = ['월세','전세','자가','기타']
sizes = [15,30,45,10]
colors = ['yellowgreen','gold','lightskyblue','lightcoral']
explode = (0,0,0.1,0)
plt.title("Pie Chart")
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
       autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis("equal")
plt.show()
```

```python
import pandas as pd
bins = [0, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]
age = titanic["age"].dropna()
data = pd.cut(age,bins,labels=labels)
data2 = data.value_counts()
plt.pie(data2,explode=explode,labels=data2.index,colors=colors,
       autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis("equal")
plt.show()
```

```python
def f(x,y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
xx,yy = np.meshgrid(x,y)
zz = f(xx,yy)

plt.title("Contour plota")
plt.contourf(xx,yy,zz,alpha=.75, cmap = "jet")
plt.contour(xx,yy,zz,colors="black")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)
xx,yy = np.meshgrid(x,y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(xx,yy,zz,rstride=1,cstride=1,cmap='hot')
plt.show()
```
